```python
import numpy as np

minmax_features: dict[str, dict[str, float]] = {}
for column in columns_without_label:
    minmax_features[column] = {
        'min': np.min(df[column]),
        'max': np.max(df[column])
    }
minmax_features
```

```python
def minmax_column(df: pd.DataFrame, column: str, minmax_features: dict[str, dict[str, float]]) -> pd.Series:
    feature_minmax = minmax_features[column]
    min_for_column = feature_minmax['min']
    max_for_column = feature_minmax['max']
    minmaxed_column = (df[column] - min_for_column) / (max_for_column - min_for_column)
    return minmaxed_column
```

```python
df_scaled = pd.DataFrame(columns=columns_without_label)
for column in columns_without_label.values:
    df_scaled[column] = minmax_column(df, column, minmax_features)
df_scaled.describe()
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_scaled, labels_resized, test_size=0.3, random_state=42, shuffle=True)
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.33, random_state=42)
```

```python
x_train.shape, y_train.shape, x_val.shape, y_val.shape, x_test.shape, y_test.shape
```

```python
features_len = len(df_scaled.columns)
features_len
```

```python
from keras import layers, Sequential

simple_model = Sequential([
    layers.InputLayer(input_shape=(features_len, )),
    layers.Dense(512, activation='relu'),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dropout(0.1),
    layers.Dense(10, activation='softmax')
])

simple_model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    patience=10,
    restore_best_weights=True
)
```

```python
history = simple_model.fit(
    x_train, y_train,
    epochs=100,
    callbacks=[early_stopping],
    validation_data=(x_val, y_val)
)
```

```python
loss, accuracy = simple_model.evaluate(x_test, y_test)
```

```python
def get_mean_and_var(feature: np.ndarray) -> tuple[float, float]:
    return np.mean(feature), np.var(feature)
```

```python
from librosa.beat import beat_track
from librosa.effects import hpss
from librosa.feature import (
    chroma_stft, rms, spectral_centroid, spectral_bandwidth,
    spectral_rolloff, zero_crossing_rate, mfcc
)

SAMPLING_RATE = 22050


def get_features_for_row(split: np.ndarray) -> np.ndarray:
    features_for_row = []

    chroma_stft_ = chroma_stft(y=split, sr=SAMPLING_RATE, hop_length=512)
    chroma_stft_mean, chroma_stft_var = get_mean_and_var(chroma_stft_)
    features_for_row.append(chroma_stft_mean)
    features_for_row.append(chroma_stft_var)

    rms_ = rms(y=split)
    rms_mean, rms_var = get_mean_and_var(rms_)
    features_for_row.append(rms_mean)
    features_for_row.append(rms_var)

    spectral_centroid_ = spectral_centroid(y=split, sr=SAMPLING_RATE)[0]
    spectral_centroid_mean, spectral_centroid_var = get_mean_and_var(spectral_centroid_)
    features_for_row.append(spectral_centroid_mean)
    features_for_row.append(spectral_centroid_var)

    spectral_bandwidth_ = spectral_bandwidth(y=split, sr=SAMPLING_RATE)
    spectral_bandwidth_mean, spectral_bandwidth_var = get_mean_and_var(spectral_bandwidth_)
    features_for_row.append(spectral_bandwidth_mean)
    features_for_row.append(spectral_bandwidth_var)

    rolloff_ = spectral_rolloff(y=split, sr=SAMPLING_RATE)[0]
    rolloff_mean, rolloff_var = get_mean_and_var(rolloff_)
    features_for_row.append(rolloff_mean)
    features_for_row.append(rolloff_var)

    zcr = zero_crossing_rate(y=split, pad=False)
    zcr_mean, zcr_var = get_mean_and_var(zcr)
    features_for_row.append(zcr_mean)
    features_for_row.append(zcr_var)

    harmony, perceptr = hpss(y=split)
    harmony_mean, harmony_var = get_mean_and_var(harmony)
    features_for_row.append(harmony_mean)
    features_for_row.append(harmony_var)

    perceptr_mean, perceptr_var = get_mean_and_var(perceptr)
    features_for_row.append(perceptr_mean)
    features_for_row.append(perceptr_var)

    tempo = beat_track(y=split, sr=SAMPLING_RATE)[0]
    features_for_row.append(tempo)

    for mfcc_ in mfcc(y=split, sr=SAMPLING_RATE):
        mfcc_mean, mfcc_var = get_mean_and_var(mfcc_)
        features_for_row.append(mfcc_mean)
        features_for_row.append(mfcc_var)
    
    return features_for_row
```

```python
from os import cpu_count
import multiprocessing as mp


def create_features_for_sample(sample: np.ndarray) -> list[list[float]]:
    sample_n_splits = len(sample) // length_of_dataset_records
    sample_splits = np.array_split(sample, sample_n_splits)
    sample_matrix = []
    with mp.Pool(cpu_count() - 1) as pool:
        for split in sample_splits:
            sample_matrix.append(pool.apply(get_features_for_row, args=(split[:length_of_dataset_records],)))
    return sample_matrix
```

```python
def minmax_test_audio_df(test_audio_df: pd.DataFrame):
    for column in test_audio_df.columns:
        feature_minmax = minmax_features[column]
        column_min = feature_minmax['min']
        column_max = feature_minmax['max']
        test_audio_df[column] = (test_audio_df[column] - column_min) / (column_max - column_min)
    return test_audio_df
```

```python
from collections import defaultdict
import heapq

most_common = defaultdict(int)

for idx, pred_sample in enumerate(prediction):
    n_largest = heapq.nlargest(3, pred_sample)
    indices_of_n_largest = [np.where(pred_sample == nth_max)[0][0] for nth_max in n_largest]
    for nth_max in indices_of_n_largest:
        most_common[labels_decoded[nth_max]] += 1

sorted(most_common.items(), key=lambda d: d[1], reverse=True)[:3]
```

```python
predicted_genres_mean = [np.mean(predicted_genre) for predicted_genre in prediction.T]
largest_means_of_predicted_genres = heapq.nlargest(3, predicted_genres_mean)
indices_of_n_largest_means = [np.where(predicted_genres_mean == nth_max)[0][0] for nth_max in largest_means_of_predicted_genres]
[(labels_decoded[idx], round(100 * largest_mean, 2)) for idx, largest_mean in zip(indices_of_n_largest_means, largest_means_of_predicted_genres)]
```

```python
sigmoided_predicted_genres_mean = 1 / (1 + np.exp(-np.array(predicted_genres_mean)))
largest_means_of_sigmoided_predicted_genres = heapq.nlargest(3, sigmoided_predicted_genres_mean)
indices_of_n_largest_sigmoided_means = [np.where(predicted_genres_mean == nth_max)[0][0] for nth_max in largest_means_of_predicted_genres]
[(labels_decoded[idx], round(100 * largest_mean, 2))
 for idx, largest_mean in zip(indices_of_n_largest_sigmoided_means, largest_means_of_sigmoided_predicted_genres)]
```
