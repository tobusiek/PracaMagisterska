```python
from pathlib import Path

test_audio_path = Path('test_audio.mp3')
test_audio_wav_path = Path('test_audio.wav')
```

```python
from pydub import AudioSegment

test_audio_segment = AudioSegment.from_mp3(test_audio_path)
test_audio_segment.export(test_audio_wav_path, format='wav')
```

```python
import librosa
import librosa.effects

test_audio, test_audio_sampling_rate = librosa.load(test_audio_wav_path)
```

```python
from IPython.display import Audio

# can crush jupyter kernel
# Audio(soad_audio, rate=soad_sampling_rate)
```

```python
test_audio_trimmed, test_audio_trim_idx = librosa.effects.trim(test_audio)
duration = 30
n_samples = int(duration * test_audio_sampling_rate)
test_audio_sample = test_audio_trimmed[:n_samples]
```

```python
Audio(test_audio_sample, rate=test_audio_sampling_rate)
```

```python
data_path = Path('data/GTZAN')
audio_path = Path(data_path.joinpath('genres_original'))
```

```python
import pandas as pd
pd.set_option('display.max_columns', None)

audio_features_30s = pd.read_csv(data_path.joinpath('features_30_sec.csv'))
audio_features_30s.head()
```

```python
import numpy as np

np.asarray(audio_features_30s.columns)
```

```python
def get_mean_and_var(feature: np.ndarray) -> tuple[float, float]:
    return np.mean(feature), np.var(feature)
```

```python
import librosa.feature

audio_length = float(len(test_audio_sample))

chroma_stft = librosa.feature.chroma_stft(y=test_audio_sample, sr=test_audio_sampling_rate)
chroma_stft_mean, chroma_stft_var = get_mean_and_var(chroma_stft)

rms = librosa.feature.rms(y=test_audio)
rms_mean, rms_var = get_mean_and_var(rms)

spectral_centroid = librosa.feature.spectral_centroid(y=test_audio_sample, sr=test_audio_sampling_rate)
spectral_centroid_mean, spectral_centroid_var = get_mean_and_var(spectral_centroid)

spectral_bandwidth = librosa.feature.spectral_bandwidth(y=test_audio_sample, sr=test_audio_sampling_rate)
spectral_bandwidth_mean, spectral_bandwidth_var = get_mean_and_var(spectral_bandwidth) 

rolloff = librosa.feature.spectral_rolloff(y=test_audio_sample, sr=test_audio_sampling_rate)
rolloff_mean, rolloff_var = get_mean_and_var(rolloff)

zero_crossing_rate = librosa.feature.zero_crossing_rate(y=test_audio_sample)
zero_crossing_rate_mean, zero_crossing_rate_var = get_mean_and_var(zero_crossing_rate)

harmony = librosa.effects.harmonic(y=test_audio_sample)
harmony_mean, harmony_var = get_mean_and_var(harmony)

# ommit those, cannot be extracted
# perceptr_mean, perceptr_var

tempo = librosa.feature.tempo(y=test_audio_sample, sr=test_audio_sampling_rate)

mfcc = librosa.feature.mfcc(y=test_audio_sample, sr=test_audio_sampling_rate)
mfcc1_mean, mfcc1_var = get_mean_and_var(mfcc[0])
mfcc2_mean, mfcc2_var = get_mean_and_var(mfcc[1])
mfcc3_mean, mfcc3_var = get_mean_and_var(mfcc[2])
mfcc4_mean, mfcc4_var = get_mean_and_var(mfcc[3])
mfcc5_mean, mfcc5_var = get_mean_and_var(mfcc[4])
mfcc6_mean, mfcc6_var = get_mean_and_var(mfcc[5])
mfcc7_mean, mfcc7_var = get_mean_and_var(mfcc[6])
mfcc8_mean, mfcc8_var = get_mean_and_var(mfcc[7])
mfcc9_mean, mfcc9_var = get_mean_and_var(mfcc[8])
mfcc10_mean, mfcc10_var = get_mean_and_var(mfcc[9])
mfcc11_mean, mfcc11_var = get_mean_and_var(mfcc[10])
mfcc12_mean, mfcc12_var = get_mean_and_var(mfcc[11])
mfcc13_mean, mfcc13_var = get_mean_and_var(mfcc[12])
mfcc14_mean, mfcc14_var = get_mean_and_var(mfcc[13])
mfcc15_mean, mfcc15_var = get_mean_and_var(mfcc[14])
mfcc16_mean, mfcc16_var = get_mean_and_var(mfcc[15])
mfcc17_mean, mfcc17_var = get_mean_and_var(mfcc[16])
mfcc18_mean, mfcc18_var = get_mean_and_var(mfcc[17])
mfcc19_mean, mfcc19_var = get_mean_and_var(mfcc[18])
mfcc20_mean, mfcc20_var = get_mean_and_var(mfcc[19])
```
