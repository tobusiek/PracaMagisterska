```python
from pathlib import Path
import pandas as pd
pd.set_option('display.max_columns', None)

data_path = Path('Data')
df = pd.read_csv(data_path.joinpath('features_3_sec.csv'))
df.head()
```

```python
df['length'].value_counts()
```

```python
length_of_dataset_records = df['length'][0]
length_of_dataset_records
```

```python
df.drop(['filename', 'length', 'perceptr_mean', 'perceptr_var'], axis=1, inplace=True)
```

```python
df[df.isna().any(axis=1)]
```

```python
columns_without_label = df.drop('label', axis=1).columns
columns_without_label
```

```python
import numpy as np

minmax_features = []
for column in columns_without_label:
    minmax_features.append((np.min(df[column]), np.max(df[column])))
minmax_features
```

```python
def minmax_column(column: pd.Series, column_id: int, minmax_features: list[tuple[float, float]]) -> pd.Series:
    min_for_column, max_for_column = minmax_features[column_id]
    minmaxed_column = (column - min_for_column) / (max_for_column - min_for_column)
    return minmaxed_column
```

```python
df_scaled = pd.DataFrame(columns=columns_without_label)
for id, column in enumerate(columns_without_label):
    df_scaled[column] = minmax_column(df[column], id, minmax_features)
df_scaled.describe()
```

```python
labels = df['label']
labels_resized = pd.get_dummies(pd.DataFrame(labels))
labels_resized
```

```python
labels_decoded = {}
column_id: int
column: str
for column_id, column in enumerate(labels_resized.columns):
    labels_decoded[column_id] = column.replace('label_', '')
labels_decoded
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(df_scaled, labels_resized, test_size=0.3, random_state=42, stratify=labels_resized)
x_val, x_test, y_val, y_test = train_test_split(x_test, y_test, test_size=0.33, random_state=42)
```

```python
x_train.shape, y_train.shape, x_val.shape, y_val.shape, x_test.shape, y_test.shape
```

```python
features_len = len(df_scaled.columns)
features_len
```

```python
from keras import layers, Sequential

model = Sequential([
    layers.InputLayer(input_shape=(features_len, )),
    layers.Dense(256, activation='relu'),
    layers.Dense(128, activation='relu'),
    layers.Dense(64, activation='relu'),
    layers.Dense(10, activation='softmax')
])

model.compile(
    optimizer='adam',
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
from keras.callbacks import EarlyStopping

early_stopping = EarlyStopping(
    patience=10,
    restore_best_weights=True
)
```

```python
history = model.fit(
    x_train, y_train,
    epochs=100,
    callbacks=[early_stopping],
    validation_data=(x_val, y_val)
)
```

```python
loss, accuracy = model.evaluate(x_test, y_test)
```

```python
import librosa

test_audio_path = Path('test_audio.wav')
test_audio, sr = librosa.load(test_audio_path)
```

```python
def get_mean_and_var(feature: np.ndarray) -> tuple[float, float]:
    return np.mean(feature), np.var(feature)
```

```python
from typing import Callable
from librosa.effects import harmonic
from librosa.feature import (
    chroma_stft, rms, spectral_centroid, spectral_bandwidth,
    spectral_rolloff, zero_crossing_rate, tempo, mfcc
)

features_without_mfcc_and_tempo = [
    chroma_stft, rms, spectral_centroid, spectral_bandwidth,
    spectral_rolloff, zero_crossing_rate, harmonic
]

def get_features_for_row(split: np.ndarray, features: Callable) -> np.ndarray:
    features_for_row = []
    for feature in features:
        feature_mean, feature_var = get_mean_and_var(feature(y=split))
        features_for_row.append(feature_mean)
        features_for_row.append(feature_var)
    features_for_row.append(tempo(y=split))
    for mfcc_ in mfcc(y=split):
        mfcc_mean, mfcc_var = get_mean_and_var(mfcc_)
        features_for_row.append(mfcc_mean)
        features_for_row.append(mfcc_var)
    return features_for_row
```

```python
len(test_audio)
```

```python
n_splits = len(test_audio) // length_of_dataset_records
test_audio_splits = np.array_split(test_audio, n_splits)
test_audio_matrix = []
```

```python
for split in test_audio_splits:
    test_audio_matrix.append(get_features_for_row(split[:length_of_dataset_records], features_without_mfcc_and_tempo))
```

```python
np.array(test_audio_matrix).shape
```

```python
test_audio_df = pd.DataFrame(test_audio_matrix, columns=columns_without_label)
test_audio_df
```

```python
def minmax_test_audio_df():
    for column_id, column in enumerate(test_audio_df.columns):
        column_min, column_max = minmax_features[column_id]
        test_audio_df[column] = (test_audio_df[column] - column_min) / (column_max - column_min)
```

```python
def convert_test_audio_df_to_float_32():
    for column in test_audio_df.columns:
        test_audio_df[column] = test_audio_df[column].astype(np.float32)
```

```python
prediction = model.predict(test_audio_df)
prediction
```

```python
from collections import defaultdict
import heapq

most_common = defaultdict(int)

for idx, pred_sample in enumerate(prediction):
    n_largest = heapq.nlargest(3, pred_sample)
    indices_of_n_largest = [np.where(pred_sample == nth_max)[0][0] for nth_max in n_largest]
    for nth_max in indices_of_n_largest:
        most_common[labels_decoded[nth_max]] += 1

sorted(most_common.items(), key=lambda d: d[1], reverse=True)[:3]
```

```python
predicted_genres_mean = [np.mean(predicted_genre) for predicted_genre in prediction.T]
largest_means_of_predicted_genres = heapq.nlargest(3, predicted_genres_mean)
indices_of_n_largest_means = [np.where(predicted_genres_mean == nth_max)[0][0] for nth_max in largest_means_of_predicted_genres]
[(labels_decoded[idx], round(100 * largest_mean, 2)) for idx, largest_mean in zip(indices_of_n_largest_means, largest_means_of_predicted_genres)]
```

```python
sigmoided_predicted_genres_mean = 1 / (1 + np.exp(-np.array(predicted_genres_mean)))
largest_means_of_sigmoided_predicted_genres = heapq.nlargest(3, sigmoided_predicted_genres_mean)
indices_of_n_largest_sigmoided_means = [np.where(predicted_genres_mean == nth_max)[0][0] for nth_max in largest_means_of_predicted_genres]
[(labels_decoded[idx], round(100 * largest_mean, 2))
 for idx, largest_mean in zip(indices_of_n_largest_sigmoided_means, largest_means_of_sigmoided_predicted_genres)]
```
